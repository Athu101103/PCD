```python

''' a. Write a program to remove left recursion (Both immediate and indirect) from the given grammar. (TO BE DONE FOR INDIRECT ) '''

'''
    Unicodes :
        ε - \u03B5

    italic_alphabet_caps = {
    'A': '𝘈', 'B': '𝘉', 'C': '𝘊', 'D': '𝘋', 'E': '𝘌', 'F': '𝘍', 'G': '𝘎',
    'H': '𝘏', 'I': '𝘐', 'J': '𝘑', 'K': '𝘒', 'L': '𝘓', 'M': '𝘔', 'N': '𝘕',
    'O': '𝘖', 'P': '𝘗', 'Q': '𝘘', 'R': '𝘙', 'S': '𝘚', 'T': '𝘛', 'U': '𝘜',
    'V': '𝘝', 'W': '𝘞', 'X': '𝘟', 'Y': '𝘠', 'Z': '𝘡'
    }

    bold_italic_alphabet_caps = {
    'A': '𝑨', 'B': '𝑩', 'C': '𝑪', 'D': '𝑫', 'E': '𝑬', 'F': '𝑭', 'G': '𝑮',
    'H': '𝑯', 'I': '𝑰', 'J': '𝑱', 'K': '𝑲', 'L': '𝑳', 'M': '𝑴', 'N': '𝑵',
    'O': '𝑶', 'P': '𝑷', 'Q': '𝑸', 'R': '𝑹', 'S': '𝑺', 'T': '𝑻', 'U': '𝑼',
    'V': '𝑽', 'W': '𝑾', 'X': '𝑿', 'Y': '𝒀', 'Z': '𝒁'
    }
'''


from collections import defaultdict

# Function to remove left recursion :
def rmLRec(G : dict) -> dict :

    # Bold italic chars.
    dashChars = {
                'A': '𝑨', 'B': '𝑩', 'C': '𝑪', 'D': '𝑫', 'E': '𝑬', 'F': '𝑭', 'G': '𝑮', 'H': '𝑯', 'I': '𝑰', 'J': '𝑱',
                'K': '𝑲', 'L': '𝑳', 'M': '𝑴', 'N': '𝑵', 'O': '𝑶', 'P': '𝑷', 'Q': '𝑸', 'R': '𝑹', 'S': '𝑺', 'T': '𝑻',
                'U': '𝑼', 'V': '𝑽', 'W': '𝑾', 'X': '𝑿', 'Y': '𝒀', 'Z': '𝒁'
                }

    newG = defaultdict(list)

    for k, v in G.items() :

        alpha_list = []
        beta_list = []

        for prod in v :

            # If there is a left recursion :
            if prod[0] == k :
                alpha_list += [prod[1 : ]]

            if alpha_list != [] and prod[0] != k:
                beta_list += [prod]

        # If left recursion occured :
        if alpha_list != [] :

            vDash = dashChars[k]

            # Including E -> TE(dash).
            for beta in beta_list :
                newG[k] += [beta + vDash]

            for alpha in alpha_list :
                newG[vDash] += [alpha + vDash]

            newG[vDash] += ['\u03B5']

        # If no left recursion occured :
        else :
            newG[k] = v


    return newG

# Main Function :
def main() :

    grammar = {'E' : ['E+T', 'T'], 'T' : ['T*F', 'F'], 'F' : ['i']}

    grammar = rmLRec(grammar)

    print('\nLeft recursive removed grammar :\n')
    for k, v in grammar.items() :
        print(k, v)
    print('\n')

if __name__ == '__main__' :
    main()
```

```python

''' d. Write a program to implement predictive parsing technique. '''


'''
    NOTE :
        1. Grammar Should be given in the reverse order(based on the depndency of non-terminals) for FIRST.
        2. Grammar can be given in any order for FOLLLOW
'''

from collections import defaultdict
from prettytable import PrettyTable

# Function to find non-terminal or not :
def isNonTerminal(V : str) :
    return V.isupper()

# Function to find first of given grammar :
def findFirst(G : dict) -> dict :

    # Resultant "first" dict.
    ans = defaultdict(list)

    for k, v in G.items() :

        for prod in v :

            # If first character is terminal, just add it :
            if not isNonTerminal(prod[0]) :
                ans[k] += [prod[0]]

            # If first character is a non-terminal :
            else :

                # We already know the first of prod[0] :
                if prod[0] in ans :
                    ans[k] += ans[prod[0]]

    return ans


# Function to find follow :
def findFollow(start : str, G : dict, first : dict) -> dict :

    # Resultant "follow" dict.
    follow = defaultdict(list)

    # Add "$" to follow of start symbol.
    follow[start] = ['$']

    """
        [Non-terminal whose follow need to be found, where it should be placed].
    """
    toBeDone = []

    for k, v in G.items() :

        for prod in v :

            indx = 0

            while indx < len(prod) :

                # In the production if current char is non-terminal :
                if isNonTerminal(prod[indx]) :

                    # If there is another character (NOT EPSILON ) :
                    if indx + 1 < len(prod) :

                        # If next character is a terminal :
                        if not isNonTerminal(prod[indx + 1]) :
                            follow[prod[indx]] += [prod[indx + 1]]
                            follow[prod[indx]] = list(set(follow[prod[indx]]))

                        # If next character is a non-terminal :
                        else :

                            # Add first of next non-terminal to follow of "prod[indx]"."
                            follow[prod[indx]] += first[prod[indx + 1]]
                            follow[prod[indx]] = list(set(follow[prod[indx]]))

                            # If first of next non-terminal contains "ε" then add follow of k to "follow" of "prod[indx]".
                            if 'ε' in first[prod[indx + 1]] :

                                if k in follow :
                                    follow[prod[indx]] += follow[k]
                                    follow[prod[indx]] = list(set(follow[prod[indx]]))
                                    toBeDone += [[k, prod[indx]]]

                                '''else :
                                    toBeDone += [[k, prod[indx]]]
                                    print('el', toBeDone)'''


                    # If next character is EPSILON,  then add follow of k to "follow" of "prod[indx]":
                    else :

                        if k in follow :
                            follow[prod[indx]] += follow[k]
                            follow[prod[indx]] = list(set(follow[prod[indx]]))
                            toBeDone += [[k, prod[indx]]]

                        '''else :
                            toBeDone += [[k, prod[indx]]]
                            print('else', toBeDone)'''

                indx += 1


    for l in toBeDone :
        follow[l[1]] += follow[l[0]]
        follow[l[1]] = list(set(follow[l[1]]))

        # Removing EPSILON :
        if 'ε' in follow[l[1]] :
            follow[l[1]].remove('ε')


    return follow


# Function to create table :
def createTable(G : dict, first : dict, follow : dict) -> dict :

    table = {}

    # Non-terminals.
    NT = list(G.keys())

    # Initializing empty dictionary for each non-terminal :
    for var in NT :
        table[var] = {}

    # Traversing the productions.
    for k, v in G.items() :

        for prod in v :

            # If the production contains only EPSILON :
            if prod == 'ε' :
                for term in follow[k] :
                    table[k][term] = prod

            # If first char of prod is terminal.
            elif not isNonTerminal(prod[0]) :
                table[k][prod[0]] = prod

            # If first char of prod is a non-terminal :
            else :

                # For each terminal in first of prod[0] set the corresponding column with the current "prod" :
                for term in first[prod[0]] :

                    if term != 'ε' :
                        table[k][term] = prod

                    else :
                        print('fkd')
                        for term in follow[k] :
                            table[k][term] = prod

    return table


# Function to parse the input :
def parse(table : dict, inp : str, start : str) -> list :

    t = PrettyTable(['Stack', 'Operation', 'Input'])

    indx = 0

    inp += '$'

    stack = ['$', start]

    while True :

        # FALSE case :
        if indx == len(inp) :
            return [False, []]

        # TRUE case :
        elif stack[-1] == inp[indx] == '$' :
            t.add_row([''.join(stack), 'ACCEPT', inp[indx : ]])
            return [True, t]

        # Stack[-1] == "i" and inp[indx] == "i" :
        if stack[-1] == inp[indx] :
            t.add_row([''.join(stack), 'remove', inp[indx : ]])
            stack.pop()
            indx += 1

        # Finding the operation to be done.

        if inp[indx] not in table[stack[-1]] :
            return [False, []]

        ops = table[stack[-1]][inp[indx]]

        t.add_row([''.join(stack), ops, inp[indx : ]])

        if ops == 'ε' :
            stack.pop()

        else :
            # Poping the non-terminal from stack.
            stack.pop()

            # Reversing and appending the "ops" to the stack.
            stack += list(ops[::-1])

    return t

# Main Function :
def main() :

    grammarR = {'F' : ['(E)', 'i'], '𝑻' : ['*F𝑻', '/F𝑻', 'ε'], 'T' : ['F𝑻'], '𝑬' : ['+T𝑬', '-T𝑬', 'ε'], 'E' : ['T𝑬']}
    grammarF = {'E' : ['T𝑬'], '𝑬' : ['+T𝑬', '-T𝑬', 'ε'], 'T' : ['F𝑻'], '𝑻' : ['*F𝑻', '/F𝑻', 'ε'], 'F' : ['(E)', 'i']}

    start = 'E'

    inp = 'i+i*i'

    first = findFirst(grammarR)

    print('\nFirst of Grammar :\n')
    for k, v in first.items() :
        print(k, v)
    print()

    follow = findFollow(start, grammarF, first)

    print('\nFollow :\n')
    for k, v in follow.items() :
        print(k, v)
    print()

    table : dict = createTable(grammarF, first, follow)

    print('\nTable :\n')
    for k, v in table.items() :
        print(k, v)
    print()


    res = parse(table, inp, start)

    print('\ninput :', inp)

    if res[0] == True :
        print('\nGiven string is valid.\n')
        print(res[1])
        print()

    else :
        print('\nGiven string is invalid.\n')

if __name__ == '__main__' :
    main()
```
